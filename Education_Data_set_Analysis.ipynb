Gender Distribution

```python

ax=sns.countplot(data=df, x="Gender")

ax.bar_label(ax.containers[0])
plt.title("Gender")


```

#From the chart we have analyzed that:
#The number of Male is greater that the number of Female students in that performence data set.

Parental_Education_Level

```python
ap=sns.countplot(data=df,x="Parental_Education_Level")
ap.bar_label(ap.containers[0])
plt.title("Parental_Education_Level")
```

From the chart, we can see the following observations:
1. The number of individuals with a High School level of parental education is significantly higher compared to those with College or Postgraduate education levels.
2. The second highest number of individuals have College-level parental education.


Family Income

```python
pu=sns.countplot(data=df,x="Family_Income")
pu.bar_label(pu.containers[0])
plt.title("Family_Income")
```

From the chart, we can see the following observations:
1. The number of individuals with a Low income is slightly higher compared to those with Medium & significantly higher than High level income.




Relation on Access_to _Resources.

```python
ax=df.groupby("Access_to_Resources").agg({"Hours_Studied":"mean","Attendance":"mean","Sleep_Hours":"mean","Previous_Scores":"mean","Tutoring_Sessions":"mean","Exam_Score":"mean","Physical_Activity":"mean"})
print(ax)



```

```python

sns.heatmap(ax,annot=True)


```

From this heatchart we can conclude that there is no/neglegible impacts on the Access to Resources 

```python
sns.boxplot(data=df,x="Exam_Score")
plt.title("Exam_Score")
```

1.Distribution piechart plan for "Access_to_Resources" Data 

```python
print(df["Access_to_Resources"].unique())
```

```python
High=df.loc[(df["Access_to_Resources"]=="High")].count()
Medium=df.loc[(df["Access_to_Resources"]=="Medium")].count()
Low=df.loc[(df["Access_to_Resources"]=="Low")].count()


list1=[High["Access_to_Resources"],Medium["Access_to_Resources"],Low["Access_to_Resources"]]
mark=["High","Medium","Low"]

plt.pie(list1,labels=mark,autopct="%1.2f%%")
plt.title("Access_to_Resources")
plt.show()
```
